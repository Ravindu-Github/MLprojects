```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_datasset=sklearn.datasets.load_boston()
house_price_datasset
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing
```

```python
houseprice_dataframe=pd.DataFrame(housing.data,columns=housing.feature_names)
houseprice_dataframe
```

```python
houseprice_dataframe['Price']=housing.target

houseprice_dataframe
```

```python
houseprice_dataframe.shape
```

```python
#checking missing values

houseprice_dataframe.isnull().sum()
```

```python
houseprice_dataframe.describe()
```

```python
#understanding the correlation

correlation=houseprice_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')

#cbar -- color bar in the rigjt side
#fmt -- how many decimal points needed
```

```python
#spliiting the data

X=houseprice_dataframe.drop(['Price'],axis=1)
Y=houseprice_dataframe['Price']
```

```python
#Training and testing data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

model selection

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction=model.predict(X_train)
training_data_prediction
```

```python
#R squared

score1=metrics.r2_score(Y_train,training_data_prediction)

#mean absoliute error

score2=metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print(score1)
```

```python
print(score2)
```

prediction on yest data

```python
testing_data_prediction=model.predict(X_test)
testing_data_prediction
```

```python
#R squared

score1=metrics.r2_score(Y_test,testing_data_prediction)

#mean absoliute error

score2=metrics.mean_absolute_error(Y_test,testing_data_prediction)

print(score1)
print(score2)
```
